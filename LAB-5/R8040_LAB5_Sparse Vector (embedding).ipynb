### 1 (a) Consider the table of term frequencies for 3 documents denoted Doc1, Doc2, Doc3 in the given table. Compute the tf-idf weights for the terms car, auto, insurance, best, for each document, using the idf values and calculate the score for any user query q. [example query: “car insurance”, “best car” ]

### (b) Apply Euclidean normalization to the tf values from the question 1 (a) and generate a table similar to the one given below.

```python
import numpy as np

# -----------------------------
# DATA FROM THE QUESTION
# -----------------------------
terms = ["car", "auto", "insurance", "best"]

# Term Frequencies (tf)
tf = {
    "car": [27, 4, 24],
    "auto": [3, 33, 0],
    "insurance": [0, 33, 29],
    "best": [14, 0, 17]
}

# Inverse Document Frequency (idf)
idf = {
    "car": 1.65,
    "auto": 2.08,
    "insurance": 1.62,
    "best": 1.50
}

# -----------------------------
# FUNCTIONS
# -----------------------------

def compute_tfidf(tf, idf):
    """Compute tf-idf table."""
    tfidf = {}
    for term in terms:
        tfidf[term] = [round(freq * idf[term], 3) for freq in tf[term]]
    return tfidf

def print_table(data, title):
    """Print table neatly."""
    print(f"\n{title}")
    print(f"{'Term':<12}{'Doc1':>10}{'Doc2':>10}{'Doc3':>10}")
    for term in terms:
        values = data[term]
        print(f"{term:<12}{values[0]:>10}{values[1]:>10}{values[2]:>10}")

def query_score(tfidf, query):
    """Calculate document score for a given query."""
    query_terms = query.lower().split()
    scores = [0, 0, 0]  # Doc1, Doc2, Doc3
    for term in query_terms:
        if term in tfidf:
            for i in range(3):
                scores[i] += tfidf[term][i]
    return scores

def euclidean_normalization(tf):
    """Apply Euclidean normalization to term frequencies."""
    norm_tf = {}
    # Compute norms for each document
    norms = []
    for doc_idx in range(3):
        norm_val = np.sqrt(sum(tf[term][doc_idx]**2 for term in terms))
        norms.append(norm_val)

    # Normalize
    for term in terms:
        norm_tf[term] = [round(tf[term][i] / norms[i], 2) if norms[i] != 0 else 0 for i in range(3)]
    return norm_tf

# -----------------------------
# MENU-DRIVEN EXECUTION
# -----------------------------
while True:
    print("\n--- TF-IDF & Normalization Calculator ---")
    print("1. Show TF table")
    print("2. Show IDF table")
    print("3. Compute and Show TF-IDF table")
    print("4. Query Document Scores")
    print("5. Show Euclidean Normalized TF table")
    print("6. Exit")
    
    choice = input("Enter your choice: ").strip()
    
    if choice == '1':
        print_table(tf, "Term Frequency Table")
        
    elif choice == '2':
        print("\nInverse Document Frequency (IDF) Table")
        print(f"{'Term':<12}{'idf':>10}")
        for term in terms:
            print(f"{term:<12}{idf[term]:>10}")
    
    elif choice == '3':
        tfidf = compute_tfidf(tf, idf)
        print_table(tfidf, "TF-IDF Table")
    
    elif choice == '4':
        tfidf = compute_tfidf(tf, idf)
        query_str = input("Enter query (e.g., 'car insurance'): ").strip()
        scores = query_score(tfidf, query_str)
        print(f"\nScores for query '{query_str}':")
        for i, score in enumerate(scores, start=1):
            print(f"Doc{i}: {round(score, 3)}")
    
    elif choice == '5':
        norm_tf = euclidean_normalization(tf)
        print_table(norm_tf, "Euclidean Normalized TF Table")
    
    elif choice == '6':
        print("Exiting program.")
        break
    
    else:
        print("Invalid choice! Please try again.")
```

### Program 2

Cosine Similarity - Compute the nearest neighbours of a word in a vector space using cosine score. <br>
(a.) Collect N documents as input. <Br>
(b.) Generate a matrix of word embeddings.[Simple count based sparse vectors & tfidf vectors] <br>
(c.) Plot a few words to understand the similarity based on the word embedding. <Br>
(d.) Compute the nearest word of any given word using cosine scores. <Br>
(e.) Plot all words nearest to a given word x. <br>
Eg: man : king = woman : X 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.decomposition import PCA
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer, ENGLISH_STOP_WORDS

# -----------------------------
# FUNCTIONS
# -----------------------------

def build_embeddings(docs, method="count"):
    """Generate word embeddings using count or tfidf, removing stopwords."""
    if method == "count":
        vectorizer = CountVectorizer(stop_words='english')  # remove stopwords
    elif method == "tfidf":
        vectorizer = TfidfVectorizer(stop_words='english')  # remove stopwords
    else:
        raise ValueError("Method must be 'count' or 'tfidf'")
    
    X = vectorizer.fit_transform(docs)
    vocab = vectorizer.get_feature_names_out()
    # Convert to word vectors: rows=words, columns=docs
    word_vectors = pd.DataFrame(
        X.toarray().T,
        index=vocab,
        columns=[f"Doc{i+1}" for i in range(len(docs))]
    )
    return word_vectors, vocab


def plot_words(word_vectors, words):
    """Plot given words in 2D space using PCA."""
    pca = PCA(n_components=2)
    reduced = pca.fit_transform(word_vectors.loc[words])
    plt.figure(figsize=(6,6))
    for i, word in enumerate(words):
        plt.scatter(reduced[i,0], reduced[i,1])
        plt.text(reduced[i,0]+0.01, reduced[i,1]+0.01, word)
    plt.title("Word Embedding Visualization (PCA)")
    plt.show()

def nearest_words(word_vectors, target_word, top_n=5):
    """Find nearest words using cosine similarity."""
    if target_word not in word_vectors.index:
        print(f"'{target_word}' not found in vocabulary.")
        return
    sims = cosine_similarity([word_vectors.loc[target_word]], word_vectors)[0]
    sim_scores = list(zip(word_vectors.index, sims))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    return sim_scores[1:top_n+1]  # exclude the word itself

def analogy(word_vectors, w1, w2, w3, top_n=1):
    """
    Solve analogy: w1 : w2 = w3 : ?
    Find w4 such that vec(w4) ≈ vec(w2) - vec(w1) + vec(w3)
    """
    for word in [w1, w2, w3]:
        if word not in word_vectors.index:
            print(f"'{word}' not in vocabulary.")
            return
    target_vec = word_vectors.loc[w2] - word_vectors.loc[w1] + word_vectors.loc[w3]
    sims = cosine_similarity([target_vec], word_vectors)[0]
    sim_scores = list(zip(word_vectors.index, sims))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    # Exclude original words
    filtered = [w for w in sim_scores if w[0] not in [w1, w2, w3]]
    return filtered[:top_n]



# -----------------------------
# MENU-DRIVEN PROGRAM
# -----------------------------
docs = []
word_vectors = None
vocab = None

while True:
    print("\n--- Word Embedding & Cosine Similarity ---")
    print("1. Input Documents")
    print("2. Build Count-based Embeddings")
    print("3. Build TF-IDF-based Embeddings")
    print("4. Plot Selected Words")
    print("5. Find Nearest Words")
    print("6. Analogy Task (man:king = woman:X)")
    print("7. Exit")
    choice = input("Enter choice: ").strip()
    
    if choice == '1':
        N = int(input("Enter number of documents: "))
        docs = [input(f"Doc{i+1}: ") for i in range(N)]
    
    elif choice == '2':
        if not docs:
            print("Please input documents first (option 1).")
            continue
        word_vectors, vocab = build_embeddings(docs, method="count")
        print("\nCount-based Embedding Matrix:")
        print(word_vectors)
    
    elif choice == '3':
        if not docs:
            print("Please input documents first (option 1).")
            continue
        word_vectors, vocab = build_embeddings(docs, method="tfidf")
        print("\nTF-IDF-based Embedding Matrix:")
        print(word_vectors)
    
    elif choice == '4':
        if word_vectors is None:
            print("Please build embeddings first (option 2 or 3).")
            continue
        words = input("Enter words to plot (comma-separated): ").lower().split(",")
        words = [w.strip() for w in words if w.strip() in vocab]
        if not words:
            print("No valid words to plot.")
            continue
        plot_words(word_vectors, words)
    
    elif choice == '5':
        if word_vectors is None:
            print("Please build embeddings first (option 2 or 3).")
            continue
        word = input("Enter target word: ").strip().lower()
        neighbours = nearest_words(word_vectors, word, top_n=5)
        if neighbours:
            print(f"Nearest words to '{word}':")
            for w, score in neighbours:
                print(f"{w} ({round(score,3)})")
            # Optional plot
            plot_words(word_vectors, [word] + [w for w,_ in neighbours])
    
    elif choice == '6':
        if word_vectors is None:
            print("Please build embeddings first (option 2 or 3).")
            continue
        w1 = input("Enter word1: ").strip().lower()
        w2 = input("Enter word2: ").strip().lower()
        w3 = input("Enter word3: ").strip().lower()
        result = analogy(word_vectors, w1, w2, w3, top_n=1)
        if result:
            print(f"{w1} : {w2} = {w3} : {result[0][0]}")
    
    elif choice == '7':
        print("Exiting program.")
        break
    else:
        print("Invalid choice! Please try again.")
```

### Program 3

Pointwise Mutual Information (PMI): Use PMI when low co-occurances of words exists. Compute the PMI instead of tf and tfidf, and recalculate the score for similarity check. <br>
[refer the link for PMI understanding: Pointwise mutual information (PMI) in NLP (listendata.com)]

```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from collections import Counter
import itertools

# -----------------------------
# FUNCTIONS
# -----------------------------

def tokenize(text):
    """Simple tokenizer: lowercase split on spaces (can be improved)."""
    return text.lower().split()

def build_pmi_matrix(docs, window_size=2):
    """
    Build PMI matrix for given documents.
    PMI(wi, wj) = log( p(wi, wj) / (p(wi)*p(wj)) )
    """
    # Tokenize all docs
    tokenized_docs = [tokenize(doc) for doc in docs]
    all_tokens = list(itertools.chain.from_iterable(tokenized_docs))
    vocab = sorted(set(all_tokens))
    
    # Unigram counts
    unigram_counts = Counter(all_tokens)
    total_tokens = sum(unigram_counts.values())
    
    # Co-occurrence counts
    co_counts = Counter()
    for tokens in tokenized_docs:
        for i, w in enumerate(tokens):
            # context window
            start = max(0, i - window_size)
            end = min(len(tokens), i + window_size + 1)
            for j in range(start, end):
                if i != j:
                    co_counts[(w, tokens[j])] += 1
    
    # Probabilities
    p_w = {w: unigram_counts[w] / total_tokens for w in vocab}
    total_co = sum(co_counts.values())
    p_wi_wj = {pair: co_counts[pair] / total_co for pair in co_counts}
    
    # Build PMI matrix
    pmi_matrix = pd.DataFrame(0.0, index=vocab, columns=vocab)
    for (wi, wj), p_ij in p_wi_wj.items():
        denom = p_w[wi] * p_w[wj]
        if p_ij > 0 and denom > 0:
            pmi = np.log2(p_ij / denom)
            if pmi > 0:  # Positive PMI
                pmi_matrix.loc[wi, wj] = pmi
    
    return pmi_matrix, vocab

def nearest_words(pmi_matrix, target_word, top_n=5):
    """Find nearest words using cosine similarity on PMI vectors."""
    if target_word not in pmi_matrix.index:
        print(f"'{target_word}' not found in vocabulary.")
        return
    sims = cosine_similarity([pmi_matrix.loc[target_word]], pmi_matrix)[0]
    sim_scores = list(zip(pmi_matrix.index, sims))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    return sim_scores[1:top_n+1]  # exclude the target word itself

# -----------------------------
# MENU-DRIVEN PROGRAM
# -----------------------------
docs = []
pmi_matrix = None
vocab = None

while True:
    print("\n--- PMI & Word Similarity ---")
    print("1. Input Documents")
    print("2. Build PMI Matrix")
    print("3. Display PMI Matrix")
    print("4. Find Nearest Words (PMI-based cosine similarity)")
    print("5. Exit")
    choice = input("Enter choice: ").strip()
    
    if choice == '1':
        N = int(input("Enter number of documents: "))
        docs = [input(f"Doc{i+1}: ") for i in range(N)]
    
    elif choice == '2':
        if not docs:
            print("Please input documents first (option 1).")
            continue
        pmi_matrix, vocab = build_pmi_matrix(docs)
        print("\nPMI Matrix built successfully.")
    
    elif choice == '3':
        if pmi_matrix is None:
            print("Please build PMI matrix first (option 2).")
            continue
        print("\nPMI Matrix (Positive PMI values only):")
        print(pmi_matrix.round(3))
    
    elif choice == '4':
        if pmi_matrix is None:
            print("Please build PMI matrix first (option 2).")
            continue
        word = input("Enter target word: ").strip().lower()
        neighbours = nearest_words(pmi_matrix, word, top_n=5)
        if neighbours:
            print(f"Nearest words to '{word}':")
            for w, score in neighbours:
                print(f"{w} ({round(score,3)})")
    
    elif choice == '5':
        print("Exiting program.")
        break
    else:
        print("Invalid choice! Please try again.")
```
